```python
# the first statistics is mean
homelessness["individuals"].mean()
#other function median(), mode(), min(), max(), var(), std(), sum(), quantile()
```

```python
#lets create a function to create the 30 percentile of a column
#we will the agg method
def pct30(column):
  return column.quantile(0.3)
```

```python
homelessness["individuals"].agg(pct30)
```

```python
#agg can also be used in more than 2 column
homelessness[["individuals","family_members"]].agg(pct30)
```

```python
#now what if we pased the multiple function?
def pct40(column):
  return column.quantile(0.4)
```

```python
homelessness["individuals"].agg([pct30,pct40])
```

```python
#now what is we want to calculate cummulative calculation?
homelessness["individuals"]
```

```python
homelessness["individuals"].cumsum() #cumsum will return an entire column rather than a single number
```

```python
#pass multiple function
def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)

print(sales[["temperature_c", "fuel_price_usd_per_l", "unemployment"]].agg([iqr, np.median]))
```

# Now lets learn how to summarize categorical data!

```python
sales.head()
```

```python
# drop duplicated in python
#you can drop duplicate based on 1 column or 2 column

#example droping 1 column
#vet_visit.drop_duplicates(subset="name")

#ex droping 2 column
#vet_visit.drop_duplicates(subset=["name","breed"])
```

```python
store_types = sales.drop_duplicates(subset=["store","type"])
print(store_types.head())
```

```python
#lets count the categorical data
sales["type"].value_counts()
```

```python
#sort the counting
sales["type"].value_counts(sort=True)
```

```python
#we can use normalized to see the propotion of the data
sales["type"].value_counts(normalize=True) #91% of the data id type A
```

```python
#how about summaries by group?
sales[sales['type'] == 'A']["weekly_sales"].mean()
```

```python
#better way you can use group function!
sales.groupby("store")["weekly_sales"].mean()
```

```python
#multiple group summaries
sales.groupby("store")["weekly_sales"].agg([min,max,sum])
```

```python
#group by multiple column
sales.groupby(["type","store"])["weekly_sales"].mean()
```

```python
#even better group by mulitple column and summarize it
sales.groupby(["type","store"])[["weekly_sales","temperature_c"]].mean()
```

```python
# Import numpy with the alias np
import numpy as np

# For each store type, aggregate weekly_sales: get min, max, mean, and median
sales_stats = sales.groupby("type")["weekly_sales"].agg([min,max,np.mean,np.median])

# Print sales_stats
print(sales_stats)

# For each store type, aggregate unemployment and fuel_price_usd_per_l: get min, max, mean, and median
unemp_fuel_stats = sales.groupby("type")[["unemployment","fuel_price_usd_per_l"]].agg([min,max,np.mean,np.median])

# Print unemp_fuel_stats
print(unemp_fuel_stats)
```

# PIVOT TABLE

```python
#remember we can use group by to calculate means
sales_mean = sales.groupby("type")["weekly_sales"].mean()
sales_mean
```

```python
#now we can use pivot function to get the same result
sales.pivot_table(values="weekly_sales", index="type") #the default function is mean
```

```python
#to pass a difference function using the aggfunc
import numpy as np

sales.pivot_table(values="weekly_sales", index="type", aggfunc= np.median)
```

```python
#pass multiple function
sales.pivot_table(values="weekly_sales", index="type", aggfunc= [np.median, np.mean])
```

```python
#remeber the group by for multiple column?
sales.groupby(["type","store"])["weekly_sales"].mean()
```

```python
sales.pivot_table(values="weekly_sales", index="store", columns="type")
```

```python
#there are several NaN becaude there are no data, lets fill this up with 0
sales.pivot_table(values="weekly_sales", index="store", columns="type", fill_value=0)
```

```python
#we can aloso put the mean of all values,BUT not include the missing values that we fill with zero
#there are several NaN becaude there are no data, lets fill this up with 0
sales.pivot_table(values="weekly_sales", index="store", columns="type", fill_value=0, margins=True)
```

# Working with pivot table

```python
#pivot table are just dataframe with sorted index
#thats way loc is ideal for subsetting pivot table!

#lets make pivot table
sales_p = sales.pivot_table(values="weekly_sales", index="date", columns="is_holiday", sort=True)
print(sales_p)
```

```python
# now you can use loc to slice data
sales_p.loc["2010-02-05":"2010-02-26"]
```

```python
#you can also calculate mean accross rows using axis parameter for each column
sales_p.loc["2010-02-05":"2010-02-26"].mean(axis="index")
```

```python
#now to calucate for each rows accross column
sales_p.loc["2010-02-05":"2010-02-26"].mean(axis="columns")
```
