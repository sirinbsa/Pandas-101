# Lets move on to sorting and subsetting

```python
homelessness.shape
```

```python
#sorting data based on the individuals
homelessness.sort_values("individuals") #the default is ascending
```

```python
#desecending sorting
homelessness.sort_values("individuals", ascending =False)
```

# Subsetting Colums

```python
homelessness["region"]
```

```python
#how to select multiple columns
homelessness[["region","state"]]
```

```python
#we could also use list to select the columns
subset_columns = ["individuals","region"]
homelessness[subset_columns]
```

# Subsetting Rows

```python
#subset rows based on codition
homelessness[homelessness['individuals'] > 10000]
```

```python
#subsetting based on text data
homelessness[homelessness["region"] == "Pacific"]
```

```python
#we can also subset by date
#for example data[data["date-of_birth"] >"2022-10-10"]
```

```python
#combined multiple conditions
is_pacific = homelessness["region"] == "Pacific"
is_state = homelessness["state"] == "Alaska"

homelessness[is_pacific & is_state]
```

```python
#or in1 line of code
homelessness[(homelessness["region"] == "Pacific") & (homelessness["state"] == "Alaska")]
```

```python
#now if we want to filter on multiple values of a categorical variable, the easiest way is to use the issin method
is_pacific_or_state = homelessness["region"].isin(["Pacific","South Atlantic"])
```

```python
homelessness[is_pacific_or_state]
```

# Adding New Column

```python
homelessness.head()
```

```python
#just add the new column index
homelessness["individuals_in_K"] = homelessness["individuals"] /100
```

```python
homelessness.head()
```

Now lets do something fun! lets find the state with individual more than 100K and sort it!

```python
more_than_100k = homelessness[homelessness["individuals_in_K"] >= 100]

more_than_100K_sorted = more_than_100k.sort_values("individuals_in_K", ascending=False)
```

```python
more_than_100K_sorted[["region","state","individuals"]]
```
