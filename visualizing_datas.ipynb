# Bar plot

```python
temp.head()
```

```python
#now lets see if want to see the avg temperature in the country overall
overall_avg_in_country = temp.groupby("country")["avg_temp_c"].mean()
print(overall_avg_in_country)
```

```python
overall_avg_in_country.plot(kind="bar", title="Overall avg temperature by country")
```

```python
mexico_dailyavg_temp = temp[temp["country"]== "Mexico"][["date","avg_temp_c"]]
```

# How to handle Missing Value

in a pandas dataframe missing value are indicated with "NaN" which means not a number.

```python
#we can use isna function
temp.isna().any()
```

```python
#count the missing value
temp.isna().sum()
```

```python
#how to remove a missing value
temp.dropna()
```

```python
#other method we can replace the missing value with zero
temp.fillna(0)
```
